```python
# Read csv MariemsDataset.csv
import pandas as pd

df = pd.read_csv('MariemsDataset.csv')
```

```python
# First 5
df.head()
```

```python
# List all options for the column Quality.
df['Quality'].unique()
```

```python
# Split column Quality by ",".
df['Quality'] = df['Quality'].str.split(',')
```

```python
# Set Quality column values to lower case.
df['Quality'] = df['Quality'].apply(lambda x: [i.lower() for i in x])
```

```python
unique_qualities = set(item.strip() for sublist in df['Quality'] for item in sublist)
unique_qualities = sorted(unique_qualities)
unique_qualities
```

```python
def calculate_quality_prevalence(dataframe):
    # For unique_qualities, count prevalence.
    quality_counts = {quality: 0 for quality in unique_qualities}
    for index, row in dataframe.iterrows():
        for quality in row['Quality']:
            quality_counts[quality.strip()] += 1
            
    # drop Qualities that have 0 as their value.
    quality_counts = {k: v for k, v in quality_counts.items() if v > 0}
    # Sort quality_counts by value.
    sorted_quality_counts = sorted(quality_counts.items(), key=lambda item: item[1], reverse=True)
    return sorted_quality_counts

calculate_quality_prevalence(df)
```

```python
# We take rows having as Quality, only items of, if another item is not in this list, reject the row: ['functional suitability', 'performance efficiency', 'compatibility','security', 'maintainability', 'reliability']
df2 = df[df['Quality'].apply(lambda x: all(item in ['functional suitability', 'performance efficiency', 'compatibility','security', 'maintainability', 'reliability'] for item in x))]
df2
```

```python
# shows rows not in df2 from df
# df2 has the Rows with qualities selected
# df3 has the excluded rows.
df3 = df[~df.index.isin(df2.index)]
df3
```

```python
calculate_quality_prevalence(df2)
```

```python
calculate_quality_prevalence(df3)
```

```python
# df2 is of interest.
# Helper function to safely parse string arrays
import ast
def parse_classification(rating_str):
    if pd.isnull(rating_str) or rating_str.strip() == '':
        return []
    try:
        parsed = ast.literal_eval(rating_str)
        if isinstance(parsed, list):
            return parsed
    except (ValueError, SyntaxError):
        pass
    return None

# Apply the parsing
df2['parsed_classification'] = df2['classification'].apply(parse_classification)
```

```python
matched_df = df2[
    ~df2['parsed_classification']
      .astype(str)                       # ensure non-strings (NaN, numbers) become str
      .str.contains("Functional Suitability|Performance Efficiency|Reliability|Maintainability|Compatibility|Interaction Capability|Flexibility|Security|Safety", na=False)
]

matched_df = matched_df[~matched_df['parsed_classification'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
matched_df
```

```python
mis_classifications = df2[df2["parsed_classification"].isnull()]
mis_classifications
```

```python
# Exclude rows in mis_classifications and matched_df from df2
df2 = df2[~df2.index.isin(mis_classifications.index)]
df2 = df2[~df2.index.isin(matched_df.index)]
df2 = df2.dropna(subset=['classification'])
df2
```

```python
# add 1 to partial if one but not all of the Quality items are in the parsed_classification array.
# add 1 to exact if all of the Quality items are in the parsed_classification array, if empty arrays, add 1 to exact.
def compute_matches(row):
    parsed_classification = [item.lower() for item in row.parsed_classification]
    quality = [item.lower() for item in row.Quality]

    partial = any(item in parsed_classification for item in quality)
    exact = bool(quality) and all(item in parsed_classification for item in quality)

    if exact:
        partial = 0

    return {'partial': int(partial), 'exact': int(exact), 'mismatch': 0 if (partial or exact) else 1 }

# Apply the function to each row and summarize the results of all rows
df2['matches'] = df2.apply(compute_matches, axis=1)
# Summarize the results
summary = df2['matches'].apply(pd.Series).sum()
summary
```
